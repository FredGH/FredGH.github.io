1/Briefly explaination of the essential differences between regular and context-free grammars.

Regular grammars (RGs) can be represented by a regular expression or a regular grammar. A regular expression (RE) is a “standard notation for characterising text sequence” [1].  RGs are usually used for text search purpose, word processing (e.g. spell checking), tagging or chucking.  REs can be used  for searching a string. For example: ’ing$’, which means match all words ending with ‘ing’. RGs can be defined as finite-state machines (FSM), a.k.a. finite-state automata [2]. They contain sequences (‘the nice boy’), selections (‘boy | girl’) and iterations (‘nice*’). A Pinker’s word chain diagram is shown as an example in the below Figure1.  
      
The main limitations of RGs are:
i) The potential generation of infinite number of sentences. In this case, the word ‘nice’ is optional but the arrow at the top (going from right to left) shows the loop. ‘Nice’ lists back to itself any number of times, therefore the following sentence could be formed: ‘The nice boy helps the police, The nice boy helps the police’, and so on.
ii)	The word chain example also fails in the ‘if/either…or/then’ example. In English sentences, ‘either’ must be followed by ‘or’, and ‘if by then’. With above FSM the sentence: ‘Either a boy helps the police then the girl helps the firemen’ could be built. This is grammatically incorrect. FSMs have no concept or “remembering early words” [3].
iii)Furthermore, RGs are always in one of a finite number of states [3]
Consequently, it appears that using regular grammars for representing natural language is not optimal.

Contextual grammars (CGs) are a superset of RGs. They resolve some of the previously mentioned issues: 
i)	CGs allow for ‘if/then’’ and/or ‘either/or’ to be constructed grammatically correctly. This is done by allowing non-terminal symbol on the right-hand side to have materials on both sides.
ii)	 ii) CGs allows for indefinite nesting. Consequently helping with resolving the central-embedding problem. ‘If’ is repeated several times in the sentence ‘If Fred comes and if Paul comes, etc..’. 
Therefore, contextual grammars are more convenient for compositional semantics.

This is an example for contextual grammar using the regular grammar as as described in [2]

S→ Either S or S 

S→ If S then S     

S→ NP VP 

NP→ Det (Adj)* N

VP→ TV NP

Det→ a | the 

Adj→ nice

N→ girl | boy | police | firemen 

TV→ helps

2/Explanation of why text generally has to be normalised before any other language processing, and brief describtion of the following: word tokenisation, stemming, lemmatisation and sentence segmentation, giving examples.

Text normalization is the process concerned with “converting [a text] to a more convenient, and standard form” to analyze it for natural language processing [1].  The process of normalization includes three steps. The first phase is the “segmentation/tokenizing of words in running text”. The second phase focuses on normalizing word formats. The third step concentrates on “segmenting sentences in running text”. The remaining of this report details these three concepts. 
The “segmentation/tokenizing” task aims at dividing a string sequence into bags of related element types (such as words, symbols or other element types), based on a defined instance of sequence of characters; namely the “token”. A simple example is tokenizing (i.e. breaking up) the sentence “Fred likes apples” by whitespace. This produces a list of three words, namely “Fred”,”likes” and “apples,” and no the whitespace separator. The process of token selection can be tricky due to the inner working of a language. For example, in English, a whitespace could be part of a ‘long’ word. For example, ‘Sutton on Sea’ is treated as a “large word” (including the whitespace). On the contrary, verbs contraction such as “you’re” need to be broken up in two words; “you” and “are”.  Electronic messages (e.g. text messaging, email and twits hashtag conventions) embedding emoticons ( e.g. “:-)” ), email address (e.g. student@gold.ac.uk), URL (e.g. www.lemonde.fr), or hashtags similar to #FollowMe also need to be tokenized appropriately. As mentioned in [1], one of the most known tokenization standard is the “Penn Treebank tokenizer”. The algorithm “separates out clitics (i.e. ”doesn’t” becomes “does” plus “n’t”), it keeps hyphenated words together, and it separates out all punctuation. The sentence of type “Joe can’t sing, jump or dance the Waltz” would be broken up as “Joe”|”can” |”n’t” |”sing”|”,”|”jump”|”or”|”dance”|”the”|”Waltz”.

The following example, using the NLTK library, represents a simple sentence tokenization. Each word is separated in a token with no whitespace. The more complex regular expression pattern is, the better it can handle complex sentences.  

```python
txt = "The tower is tall"
pattern = r'\w+'
nltk.regexp_tokenize(txt, pattern)
```

The second task involves “normalizing word formats”. Regular expression plays an important role in standardizing words/tokens to enable analysis and comparison. For example, a usual task is to ensure all words/tokens are set to the same case (lower or upper case), a.k.a. “case folding”. In this case “the” and “The” words are considered to be equivalent.  Normalization could also involve standardization of words to help with comparison. For example, the word “Republic” could be found shortened in texts as “Rep”. Therefore, “Rep.” needs to be mapped to “Republic”. Although these two methods help with machine learning algorithm generalization, they could also skew the results. The country “US” could be mistaken for the pronoun “us”. Consequently, the advantage of normalization should be weighted versus the loss of the original spelling/meaning.         
More advanced tasks include lemmatization, stemming and sentence segmentation. Lemmatization refers to finding words’ root. For example, the conjugated verb “runs” and the past version “ran” have a common “lemma” (i.e. are forms of the verb “run”). A “lemmatizer” is a process that maps words to its root. As explained in [1], sophisticated method in lemmatization consists of parsing words in their stem (the central morpheme) and their affixes (i.e. prefix/suffix). For examples, the word ‘dogs’ contain the stem ‘dog’ and the suffix ‘s’. 
Stemming is sub class of lemmatization which focuses on removing word’s suffixes. For example, the stem of “runs” is “run”. In this case “ran” would not be related to its verb root (i.e. “run”).  The below examples demonstrate the use of the NLTK built-in stemmer objects. In this simple example, all the stemmers produce the same output, at the exception of the verb ‘gives’ which is rendered as ‘give’ or ‘giv’ depending on the stemmer implementation. This is expected behaviour as the suffix extraction rules can be slightly different from one implementation to the next. More importantly, the “lemmatizer” is the only one that derives the correct root for the verb ‘drove’, i.e. ‘drive’. In all stemmer and “lemmatizer” instances, the verb ‘owed’ was correctly mapped to ‘owe’ (as expected). In all cases, ‘tickets’ was stemmed/lemmatized to ‘ticket’. ‘Additional’ was stemmed to ‘ addit’ and lemmized to 'additional'.

```python
raw = "Paul drove down the lane. Jack owed Peter money. Mark gives Peter additional tickets."
tokens = nltk.word_tokenize(raw)
porter = nltk.PorterStemmer()
[porter.stem(t) for t in tokens]
```

```python
lancaster = nltk.LancasterStemmer()
[lancaster.stem(t) for t in tokens]
```

```python
snowball = nltk.SnowballStemmer('english')
[snowball.stem(t) for t in tokens]
```

```python
wnl = nltk.WordNetLemmatizer()
[wnl.lemmatize(t, pos='v') for t in tokens]
```

Sentence segmentation refers to the task of dividing text into separate sentences, using punctuation as separation indicators (e.g. commas, periods, exclamation marks, etc.) [1]. Due to the ambiguous nature of the ‘periods’, in the case of the abbreviations such as  ‘Mr.’ or ‘Inc.’, sentence and word tokenisation can be used in parallel with machine learning to establish whether the ‘period’ relates to punctuation or abbreviation.  Depending on the language of the text under analysis, different algorithm can be used perform. A well-known algorithm for Chinese is the “MaxMatch”. As mentioned in [1], although it suitable for Chinese, it is not for English.

3/Words corpus search with NLTK for: 

a.The first 10 words that start with 'ed'

b.All words that begin with 'mother' or 'father'

c.All words more than 18 characters long that include the sequence 'art', 
e.g. 'compartmentalization'

d.All words that incude the vowels 'a', 'e', 'i', 'o', 'u' exactly once and in that order, and no other words. E.g. 'facetious'

e.All words that start with a consonant and include both 're' and 'er', in that order, e.g. 'trekker'.

As described in NLTK book 3.4, the  Words Corpus is used. A preprocess is applied to remove any proper names, 
and ensure all words are in small caps

```python
import re
wordlist = [w for w in nltk.corpus.words.words('en') if w.islower()]
```

^ed is the regular expression that retrieves all words starting with 'ed' from wordlist

[:10] selects the first 10 words of the list

```python
print("The first 10 words that start with 'ed': ")
[w for w in wordlist if re.search('^ed', w)][:10]
```

^mother|^father is the regular expression that is composed two sequences ('^mother' and '^father') and a disjunction ('|' meaning 'or')

```python
print("All words that begin with 'mother' or 'father': ")
[w for w in wordlist if re.search('^mother|^father', w)]
```

First the search returns all possible word containing the string 'art'.

Then, each word strictly greater than 18 characters is returned. 'Strictly' is an assumption. If the user wants greater or equal to 18, then '>' should be replaced by '>='

```python
print("All words more than 18 characters long that include the sequence 'art', e.g. 'compartmentalization': ")
[w for w in wordlist if re.search('art', w) and len(w)>18]
```

For each vowel bar the last one:
    
 1. The vowel is succeeded by {n,m}. This means at least 'n' repeats and no more than 'm' repeats. 
    When n=1 and m=1, this is to ensure each vowel is found exactly once.
    
 2. the '[^aeiou]' following the repeats means the next set of character should not be a vowel
 
 3. the '*' means 2. can be repeated 0 to many times.

```python
print("All words that incude the vowels 'a', 'e', 'i', 'o', 'u' exactly once and in that order, and no other words. E.g. 'facetious': ")
[w for w in wordlist 
 if re.search('a{1,1}[^aeiou]*e{1,1}[^aeiou]*i{1,1}[^aeiou]*o{1,1}[^aeiou]*u{1,1}[^aeiou]*$', w)]
```

1. ^[^(aeiou)]: ensures the word starts with a consonant
    
2. .*re{1,}: then it accepts any character followed by 're' repeated at least once.
 
3. .*re{1,}: then it accepts any character followed by 're' repeated at least once.

```python
print("All words that start with a consonant and include both 're' and 'er', in that order, e.g. 'trekker'")
[w for w in wordlist if re.search('^[^(aeiou)].*re{1,}.*er{1,}', w)]
```

5/Grammar Rules Definition (a few exmaples)  
1. If the girl likes sweet things then she eats cake or candy 
2. The fox runs if the dog barks at it 
3. The little black cat ran from the big angry dog. 
4. Maria gave Pedro a cake 
5. Tariq gave a cake to Elena 
6. *Tariq gave to Elena a cake 
7. *If the fox runs if the dog barks at it 
8. *The boy likes Elena a cake.


```python
print ("Start - Grammar rule definition")
grammar = nltk.CFG.fromstring("""
S -> 'either' S 'or' S
S -> 'if' S 'then' S
S -> S 'if' S 
S -> NP VP
NP ->  Det N1 | N1 
N1 -> N | Adj N1 
NP -> PN 
NP -> Pronoun
NP -> NP 'or' NP
VP -> V NPt
VP -> V PP
VP -> V AP
VP -> V
VP -> V PN NP
VP -> VP PP 
PP -> P NP
AP -> Adj N
Det -> 'a' | 'the' | epsilon 
N -> 'girl' | 'boy' | 'dog' | 'cat' | 'burgers' | 'candy' | 'cream' | 'cake' | 'things' |'fox' | 'cake'
V -> 'eats' | 'likes' | 'sat' |'runs' |'barks' |'ran' | 'gave'
P -> 'on' | 'at' | 'from' | 'to'
Det -> 'the' | 'a' 
Adv -> 'then'
Adj -> 'little' | 'sweet' | 'black' | 'angry' | 'big' 
Pronoun -> 'she' | 'it'
PN -> 'Maria' | 'Pedro'  | 'Tariq' | 'Elena'
 
""")
print ("End - Grammar definition")
```

```python
text = ['if','the','girl','likes', 'sweet', 'things', 'then','she', 'eats', 'cake', 'or','candy' ]
grammar_parser_and_tree_display(text, grammar)

text = ['the','fox','runs', 'if', 'the', 'dog', 'barks','at', 'it' ]
grammar_parser_and_tree_display(text, grammar)

text = ['the','fox','runs','if', 'the', 'dog', 'barks','at', 'it' ]
grammar_parser_and_tree_display(text, grammar)

text = ['the','little','black','cat', 'ran', 'from', 'the','big', 'angry', 'dog' ]
grammar_parser_and_tree_display(text, grammar)

text = ['Maria','gave','Pedro','a', 'cake']
grammar_parser_and_tree_display(text, grammar)

text = ['Tariq','gave','a','cake', 'to', 'Elena']
grammar_parser_and_tree_display(text, grammar)

#As expected this does not produce a result
text = ['Tariq','gave', 'to', 'Elena', 'a','cake']
grammar_parser_and_tree_display(text, grammar)

#As expected this does not produce a result
text = ['if','the', 'fox', 'runs', 'if','the', 'dog', 'barks', 'at', 'it']
grammar_parser_and_tree_display(text, grammar)

#TODO: The below is not behaving as expected -> should not produce a result
text = ['the','boy', 'likes', 'Elena', 'a','cake']
grammar_parser_and_tree_display(text, grammar)
```
